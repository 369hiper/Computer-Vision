```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

# Read in the Image
image = cv2.imread('C:/Users/nsida/Downloads/Computer-Vision-with-Python/DATA/Brain_MRI.jpeg')
```

```python
image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
#Convert the Image to GrayScale
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
# Creating a Custom kERNEL
#3x3 array for edge detection
sobel_x = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])
#Perform convolution using filter2D, which has inputs: (grayscale image)
#Bitdepth & kernel
filtered_img = cv2.filter2D(gray, -1, sobel_x)
plt.imshow(filtered_img, cmap='gray')
```

```python
#Now we're going to apply the Gaussian Blur Filter
gray_blur = cv2.GaussianBlur(gray, (5,5), 0)
# Now we'll plot these images side by side
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.set_title('Sobel_X')
ax1.imshow(filtered_img, cmap='gray')

ax2.set_title('Gaussian Blur')
ax2.imshow(gray_blur, cmap='gray')
```

# Test Performance with a High Pass Filter

```python
# High Pass Filter

# 3x3 sobel filter for edge Detection

sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])
# Filter the orignal and Blurred Grayscale image using Filter2D
filtered = cv2.filter2D(gray_blur, -1, sobel_x)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.set_title('Sobel_X')
ax1.imshow(filtered, cmap='gray')

ax2.set_title('Gaussian Blur')
ax2.imshow(filtered_img, cmap='gray')
```
