```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("C:/Users/nsida/Downloads/Computer-Vision-with-Python/DATA/00-puppy.jpg")
```

```python
plt.imshow(image)
```

```python
blur = cv2.GaussianBlur(image, (155,441),0)
plt.imshow(blur) 
```

```python
kernel = np.ones((15,15), 'uint8')
```

```python
dilate = cv2.dilate(image,kernel, iterations=1)
erode = cv2.erode(image,kernel,iterations=1)
```

```python
plt.imshow(dilate)
```

```python
plt.imshow(erode)
```
